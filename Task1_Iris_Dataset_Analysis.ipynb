```python
# scatter plot
```

```python
plt.figure(figsize=(8,5))

plt.scatter(
    iris['sepal_length'],
    iris['petal_length']
)

plt.title("Sepal Length vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")

plt.show()
```

# # insight is that:  Different iris species show distinct patterns between sepal length and petal length.

```python
# Histrogram
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    iris['sepal_length'],
    bins=10,
    kde=True
)

plt.title("Distribution of Sepal Length")
plt.show()
```

# insight: Sepal length follows an approximately normal distribution.

 # Box plot

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    x='species',
    y='petal_length',
    data=iris
)

plt.title("Petal Length by Species")
plt.show()
```
